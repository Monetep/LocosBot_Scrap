Tambien se puede buscar por los selectores CSS.
Los selectores son un metodo para definir los bloques de HTML a los cuales queres asignar un estilo.
Se pueden usar de la siguiente manera:
    -Por ejemplo si llamo a soup.select("p a") estoy solicitando todas las tags a dentro de todos los tags p.
    -Si hago soup.select("body p a"), busco las tags a dentro de p dentro de body.
    -Con soup.select("p.outer-text") busco todas las tags p que tengan clase outer-text.
    -Coun soup.select("p#first") busco todas las tags p con el id= "first".    
    -Con soup.select(".outer-text") busco todas las clases que se llaman outer-text.
El método select de soup, nos devuelve una lista de soup, como find y findall.      

```python
site= "http://gameranx.com/updates/id/159332/article/grand-theft-auto-v-tops-ps4-digital-sales-rocket-league-takes-second-place/"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page, 'html.parser')
print(soup.prettify())
```

Ejemplo usando una pagina del clima de USA.
Vamos a obtener cual es el id, y luego la clase que tiene los datos del clima.
Para eso usamos la funcionalidad de inspeccionar del chrome.
Se seleccionan las clases que tienen los datos que necesitamos.
Y luego inspeccionamos la imagen y obtenemos el nombre alt que le pusieron a la imagen para obtener todo el nombre alternativo que le pusieron que en este caso es bastante explicativo.

```python
site= "http://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168"
hdr = {'User-Agent': 'Mozilla/5.0'}
req = Request(site,headers=hdr)
page = urlopen(req)
soup = BeautifulSoup(page, 'html.parser')
seven_day = soup.find(id="seven-day-forecast")
forecast_items = seven_day.find_all(class_="tombstone-container")
tonight = forecast_items[0]
print(seven_day.prettify())
```

```python
period = tonight.find(class_="period-name").get_text()
short_desc = tonight.find(class_="short-desc").get_text()
temp = tonight.find(class_="temp").get_text()
print(period)
print(short_desc)
print(temp)
```

```python
img = tonight.find("img")
desc = img['title']
print(desc)
```

Podemos hacer los mismos pasos de recien pero combinando la seleccion de clases, tags y css selectors.
En este caso lo que se hace es recorrer el html buscando todas las clases que se llamen tombstone-container, y que adentro tengan la clase period-name, de 

```python
period_tags = seven_day.select(".tombstone-container .period-name")
periods = [pt.get_text() for pt in period_tags]
periods
```

Se aplica un concepto similar en el siguiente ejemplo, en el cual recorremos todo el html, pero tomando unicamente los tags que me interesan, para estos casos lo insertamos todo en una lista.

```python
short_descs = [sd.get_text() for sd in seven_day.select(".tombstone-container .short-desc")]
temps = [t.get_text() for t in seven_day.select(".tombstone-container .temp")]
descs = [d["title"] for d in seven_day.select(".tombstone-container img")]

print(short_descs)
print(temps)
print(descs)
```

Ahora vamos a utilizar todo lo visto, pero lo vamos a combinar con la libreria pandas para manejar dataframes
Primero insertamos todas las listas que hicimos antes, de forma ordenada en un dataframe.

```python
import pandas as pd
weather = pd.DataFrame({
        "period": periods, 
        "short_desc": short_descs, 
        "temp": temps, 
        "desc":descs
    })
weather
```

El metodo pd.str.extract lo que hace es a un valor obtener cierto otro haciendo uso de una regular expression.
Con ?P<temp_num> le doy un valor a lo que voy a extraer, con \d+ me quedo con todos los valores numericos, si le pusiera una sola \d tomaria el primer caracter de un valor numerico.

```python
temp_nums = weather["temp"].str.extract("(?P<temp_num>\d+)", expand=False)
weather["temp_num"] = temp_nums.astype('int')
temp_nums
```

```python
weather["temp_num"].mean()
```

Nos quedamos unicamente con los valores que suceden durante la noche.

```python
is_night = weather["temp"].str.contains("Low")
weather["is_night"] = is_night
is_night
```
